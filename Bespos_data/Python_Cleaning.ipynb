```python
# Importing necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Getting the data
product_data= pd.read_csv('Dataset/PRODUCT.csv')
product_batch_data= pd.read_csv('Dataset/PRODUCT_BATCH.csv')
transaction_data= pd.read_csv('Dataset/TRANSACTION.csv')
```

```python
# Clean product_data dataset
product_data.dropna(subset=['name'], inplace=True)
product_data['name'] = product_data['name'].str.upper()
product_data['name'] = product_data['name'].str.strip()
product_data['name'] = product_data['name'].apply(lambda x: x if x != 'NAN' else np.nan)
product_data.dropna(subset=['name'], inplace=True)
product_data.reset_index(drop=True, inplace=True)
```

```python
# Drop column other than "id,unit_id,name,product_type_id,product_measurement_unit_id,product_manufacture_id,product_brand_id,product_rack_id,is_expiry,max_discount_percentage,sales_price,expiry_days" from table product_data

product_data.drop(columns=[col for col in product_data.columns if col not in ['id','unit_id','name','product_type_id','product_measurement_unit_id','product_manufacture_id','product_brand_id','product_rack_id','is_expiry','default_discount_percentage','max_discount_percentage','is_stock_data','is_consumable','is_nil_stock','sales_price','expiry_days']], inplace=True)
```
